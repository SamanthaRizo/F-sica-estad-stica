```julia
function Caminata(n) # n es la cantidad de elementos de la lista
    Caminante=[] # Definimos un arreglo de ceros
    for c in 1:n
        Pasos=[] # Definimos un arreglo de ceros
        for p in 1:n
            push!(Pasos,rand((-1,1))) # Se insertan los elementos de rand((-1,1) en la lista Pasos
        end
        push!(Caminante,Pasos) # Se insertan los elementos de Pasos en la colección Caminante
    end
    return Caminante
end
```

#### Ejercicio 4. (Nota: El código debe tener como variables a n y L). Construye ahora L = 10, 000 listas distintas para cada uno de los siguientes valores de n: $n = {10, 100, 1000, 10000}$. (¡No las reportes!; n son el número de realizaciones o pasos). Gráfica el promedio de cada una de las listas como función del número de lista, fijando la escala de los ejes en [-1,1] en todas las gráficas. 

```julia
function Lista_Caminantes_Pasos(n,L) # n es la cantidad de pasos
                                     # L son la cantidad de listas para cada valor de n
    Caminata_L=[] #Definimos un arreglo de ceros
    for l in 1:L
        Pasos_n=[] #Definimos un arreglo de ceros
        for N in 1:n
            push!(Pasos_n,rand((-1,1))) # Se insertan los elementos de rand((-1,1) en la lista Pasos_n
        end
        push!(Caminata_L,Pasos_n)  # Se insertan los elementos de Pasos_n en la colección Caminata_L
    end
    return Caminata_L
end
```

```julia
function promedio_Caminata_Pasos(Caminata_L)
    promedios=[] # Definimos un arreglo de ceros
    for i in 1:length(Caminata_L) 
        append!(promedios,mean(Caminata_L[i]))
    end
    return promedios
end
```

```julia
L = 10000
n=[10,100,1000,10000]
n10 = Lista_Caminantes_Pasos(10,L)
n100 = Lista_Caminantes_Pasos(100,L)
n1000 = Lista_Caminantes_Pasos(1000,L)
n10000 = Lista_Caminantes_Pasos(10000,L);
```

```julia
prom10=promedio_Caminata_Pasos(n10)
prom100=promedio_Caminata_Pasos(n100)
prom1000=promedio_Caminata_Pasos(n1000)
prom10000=promedio_Caminata_Pasos(n10000);
```

```julia
plot(prom10,ylabel="Promedio", ylims = (-1.1,1.1), xlabel="Número de pasos", title="n = 10 pasos",label= false,color=:yellow)
```

```julia
plot(prom100,ylabel="Promedio", ylims = (-1.1,1.1), xlabel="Número de pasos", title="n = 100 pasos",label= false,color=:red)
```

```julia
plot(prom1000,ylabel="Promedio", ylims = (-1.1,1.1), xlabel="Número de pasos", title="n = 1000 pasos",label= false,color=:orangered)
```
