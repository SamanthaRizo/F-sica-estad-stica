```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n = 1000 
δ=1
X = np.zeros(n)
Y = np.zeros(n)
for t in range(n-1):
    X[t+1] = X[t]+np.random.normal(0,1)*np.sqrt((δ**2)*t)
    Y[t+1] = Y[t]+np.random.normal(0,1)*np.sqrt((δ**2)*t);

#Graficamos
plt.plot(X,Y,label="X(t)")
plt.title("Plano x-y")
plt.xlabel('X(t)')
plt.ylabel('Y(t)')
plt.show()
```
