```julia
using Plots
using XLSX
using DelimitedFiles
using StatsBase
using ExcelFiles
using LaTeXStrings
using DataFrames
```

```julia
function p(v,v0,γ,β,tiempo)
    probabilidad=[]
   for t in tiempo
        proba=(exp(-(v-v0*(exp(-2*γ*t)))^2/(β^2/γ*(1-(exp(-2*γ*t))))))*(2*3.1416*((β^2)/(2*γ))*(1-(exp(-2*γ*t))))^(-1/2)
        push!(probabilidad,proba)
    end
    return probabilidad
end
```

```julia
plot(title="Probabilidad para γ=1, β=1",xlabel="t",ylabel="p(v = cte,t)")
t=collect(0:0.1:10)
plot!(t,p(0,0,1,1,t),label="v = 0")
plot!(t,p(.5,0,1,1,t),label="v = 0.5")
plot!(t,p(1,0,1,1,t),label="v = 1")
plot!(t,p(2,0,1,1,t),label="v = 2")
plot!(t,p(3,0,1,1,t),label="v = 3")
plot!(t,p(4,0,1,1,t),label="v = 4")
plot!(t,p(5,0,1,1,t),label="v = 5")
plot!(t,p(10,0,1,1,t),label="v = 10")
```
